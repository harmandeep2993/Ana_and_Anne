```python
import random
import time

def the_data_scientist_simulator(code_quality):
    """
    Simulates the emotional journey of running a new Python script.
    """
    emotions = ["Pure Ego", "Confusion", "Despair", "Acceptance", "Stack Overflow"]
    
    print(f"--- Starting Analysis with {code_quality}% confidence ---")
    
    for stage in emotions:
        time.sleep(1) # Dramatic pause
        print(f"Current State: {stage}...")
        
        # If confidence is low, the chance of a 'Mental Crash' increases
        if random.randint(0, 100) > code_quality:
            return "❌ Error: Logic not found. Suggesting a coffee break."

    return "✅ It worked! (I have no idea why, don't touch anything.)"

# Example usage:
print(the_data_scientist_simulator(code_quality=15))
```
